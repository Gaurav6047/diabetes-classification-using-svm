```python
# ============================================================
# Diabetes Prediction Project (Advanced)
# ============================================================

# ==========================
# 1️⃣ Importing Dependencies
# ==========================
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score, roc_curve
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from catboost import CatBoostClassifier
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline
import pickle

# Optional: SHAP for feature explainability
import shap

# ===
```

```python
#=========================================================
# 2️⃣ Load and Explore Dataset
# ============================================================
df = pd.read_csv('diabetes.csv')

print("Dataset Preview:")
print(df.head())
print("\nDataset Info:")
print(df.info())
print("\nMissing Values:")
print(df.isnull().sum())
print("\nShape:", df.shape)
print("\nStatistical Summary:")
print(df.describe())
print("\nPregnancies Value Counts:")
print(df["Pregnancies"].value_counts())
print("\nOutcome Counts:")
print(df['Outcome'].value_counts())
# 0 -> Non-Diabetic
# 1 -> Diabetic
print("\nMean Values by Outcome:")
print(df.groupby('Outcome').mean())

```

```python
# ============================================================
# 3️⃣ Visualizations
# ============================================================
sns.countplot(x='Outcome', data=df)
plt.title('Count of Diabetic vs Non-Diabetic')
plt.show()

plt.hist(df['Age'], bins=10, color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

sns.boxplot(x='Outcome', y='Glucose', data=df)
plt.title('Glucose Level vs Outcome')
plt.show()

plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

sns.pairplot(df, vars=['Glucose','BMI','Age','Insulin'], hue='Outcome')
plt.show()

sns.histplot(df['Pregnancies'], bins=15, kde=True, color='orange')
plt.title('Distribution of Pregnancies')
plt.xlabel('Number of Pregnancies')
plt.show()

sns.scatterplot(x='Age', y='BMI', hue='Outcome', data=df)
plt.title('BMI vs Age colored by Outcome')
plt.show()

# Correlation with target
corr = df.corr()
plt.figure(figsize=(8,5))
sns.heatmap(corr[['Outcome']].sort_values(by='Outcome', ascending=False), annot=True, cmap='coolwarm')
plt.title('Feature Correlation with Outcome')
plt.show()

```

```python
# ============================================================
# 4️⃣ Data Preprocessing & Standardization
# ============================================================
X = df.drop(columns='Outcome', axis=1)
Y = df['Outcome']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X = X_scaled
print(X)
print(Y)

```

```python
# ============================================================
# 5️⃣ Train-Test Split
# ============================================================
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, stratify=Y, random_state=42
)
print("Train-Test Split Shapes:", X_train.shape, X_test.shape)
```

```python
# ============================================================
# 6️⃣ Model Training
# ============================================================
# Random Forest
rf = RandomForestClassifier()
rf_params = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5, 10]
}

# SVM with RBF Kernel
svm_model = SVC(probability=True)
svm_params = {
    'C': [0.1, 1, 10],
    'gamma': ['scale','auto'],
    'kernel': ['rbf', 'poly', 'linear']
}

# Logistic Regression
lr = LogisticRegression(max_iter=1000)
lr_params = {
    'C': [0.1, 1, 10],
    'penalty': ['l2'],
    'solver': ['lbfgs']
}

# CatBoost
cat_model = CatBoostClassifier(
    iterations=500,
    learning_rate=0.1,
    depth=6,
    verbose=100,
    random_seed=42
)

# GridSearchCV for RF, SVM, LR
grid_rf = GridSearchCV(rf, rf_params, cv=5, scoring='accuracy')
grid_rf.fit(X_train, Y_train)
print("Best RF Parameters:", grid_rf.best_params_)
print("Best RF CV Accuracy:", grid_rf.best_score_)

y_pred_rf = grid_rf.predict(X_test)
print("Random Forest Test Accuracy:", accuracy_score(Y_test, y_pred_rf))

grid_svm = GridSearchCV(svm_model, svm_params, cv=5, scoring='accuracy')
grid_svm.fit(X_train, Y_train)
print("Best SVM Parameters:", grid_svm.best_params_)
print("SVM Test Accuracy:", accuracy_score(Y_test, grid_svm.predict(X_test)))

grid_lr = GridSearchCV(lr, lr_params, cv=5, scoring='accuracy')
grid_lr.fit(X_train, Y_train)
print("Best LR Parameters:", grid_lr.best_params_)
print("Logistic Regression Test Accuracy:", accuracy_score(Y_test, grid_lr.predict(X_test)))

# CatBoost
cat_model.fit(X_train, Y_train)
print("CatBoost Test Accuracy:", accuracy_score(Y_test, cat_model.predict(X_test)))

```

```python
# ============================================================
# 7️⃣ SMOTE to balance dataset
# ============================================================
smote = SMOTE(random_state=42)
X_smote, Y_smote = smote.fit_resample(X_train, Y_train)

grid_lr.fit(X_smote, Y_smote)
grid_svm.fit(X_smote, Y_smote)
grid_rf.fit(X_smote, Y_smote)

Y_pred_lr = grid_lr.predict(X_test)
Y_pred_svm = grid_svm.predict(X_test)
Y_pred_rf = grid_rf.predict(X_test)

print("Test Accuracy on RF (SMOTE):", accuracy_score(Y_test, Y_pred_rf))
print("Test Accuracy on SVM (SMOTE):", accuracy_score(Y_test, Y_pred_svm))
print("Test Accuracy on LR (SMOTE):", accuracy_score(Y_test, Y_pred_lr))

print("\nSMOTE Results:")
print("LR--", classification_report(Y_test, Y_pred_lr))
print("SVM--", classification_report(Y_test, Y_pred_svm))
print("RF--", classification_report(Y_test, Y_pred_rf))

```

```python
# ============================================================
# 8️⃣ Feature Importance (Random Forest)
# ============================================================
importances = grid_rf.best_estimator_.feature_importances_
plt.barh(df.columns[:-1], importances)
plt.title("Feature Importance (Random Forest)")
plt.xlabel("Importance Score")
plt.show()
```

```python
# ============================================================
# 9️⃣ Confusion Matrix Function
# ============================================================
def heatmap(Y_test, y_pred, title):
    cm = confusion_matrix(Y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(title)
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

print("Logistic Regression")
heatmap(Y_test, Y_pred_lr, "Confusion Matrix - Logistic Regression")
print("SVM")
heatmap(Y_test, Y_pred_svm, "Confusion Matrix - SVM")
print("Random Forest")
heatmap(Y_test, Y_pred_rf, "Confusion Matrix - Random Forest")

```

```python
# ============================================================
#  🔟 Predictive Model (Single Input)
# ============================================================
input_data = (5,166,72,19,175,25.8,0.587,51)
input_data_as_numpy_array = np.asarray(input_data).reshape(1,-1)
std_data = scaler.transform(input_data_as_numpy_array)

prediction = grid_lr.predict(std_data)
print("\nSingle Prediction:", "Diabetic" if prediction[0]==1 else "Non-Diabetic")

```

```python
# ============================================================
# 1️⃣1️⃣ Save Trained Pipeline
# ============================================================
pipeline = ImbPipeline(steps=[
    ('scaler', StandardScaler()),
    ('smote', SMOTE(random_state=42)),
    ('model', grid_lr)
])
pipeline.fit(X_train, Y_train)

with open("diabetes_pipeline.pkl", "wb") as f:
    pickle.dump(pipeline, f)

print("✅ Pipeline saved as diabetes_pipeline.pkl")

# ============================================================
# 1️⃣2️⃣ Load & Test Pipeline
# ============================================================
with open("diabetes_pipeline.pkl", "rb") as f:
    loaded_pipeline = pickle.load(f)

sample = [6,148,72,35,0,33.6,0.627,50]
prediction = loaded_pipeline.predict([sample])[0]
print("Loaded Pipeline Prediction:", "Diabetic" if prediction==1 else "Non-Diabetic")

# ============================================================
# 1️⃣3️⃣ Optional: SHAP Feature Explainability
# ============================================================
explainer = shap.TreeExplainer(grid_rf.best_estimator_)
shap_values = explainer.shap_values(X_test)
shap.summary_plot(shap_values, X_test, feature_names=df.columns[:-1])
```
