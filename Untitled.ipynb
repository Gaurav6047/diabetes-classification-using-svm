## Train Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42, stratify=Y
)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from catboost import CatBoostClassifier
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline
```

```python
# Step 7: Define Pipelines and Hyperparameters

# 7a. Logistic Regression
pipe_lr = Pipeline([
    ('scaler', StandardScaler()),
    ('smote', SMOTE(random_state=42)),
    ('classifier', LogisticRegression(max_iter=1000))
])

lr_params = {
    'classifier__C': [0.1, 1, 10],
    'classifier__penalty': ['l2'],
    'classifier__solver': ['lbfgs']
}

# 7b. Random Forest
pipe_rf = Pipeline([
    ('scaler', StandardScaler()),
    ('smote', SMOTE(random_state=42)),
    ('classifier', RandomForestClassifier())
])

rf_params = {
    'classifier__n_estimators': [50, 100, 200],
    'classifier__max_depth': [None, 5, 10],
    'classifier__min_samples_split': [2, 5, 10]
}

# 7c. SVM
pipe_svm = Pipeline([
    ('scaler', StandardScaler()),
    ('smote', SMOTE(random_state=42)),
    ('classifier', SVC(probability=True))
])

svm_params = {
    'classifier__C': [0.1, 1, 10],
    'classifier__gamma': ['scale', 'auto'],
    'classifier__kernel': ['rbf', 'poly', 'linear']
}

# 7d. CatBoost
pipe_cat = Pipeline([
    ('smote', SMOTE(random_state=42)),
    ('classifier', CatBoostClassifier(verbose=0))
])

cat_params = {
    'classifier__depth':[4,6,8],
    'classifier__learning_rate':[0.01,0.05,0.1],
    'classifier__iterations':[200,500]
}

```

```python
# Step 8: Train Models using GridSearchCV

grid_lr = GridSearchCV(pipe_lr, lr_params, cv=5, scoring='accuracy')
grid_lr.fit(X_train, Y_train)
print("Best LR Parameters:", grid_lr.best_params_)
evaluate_model(grid_lr.best_estimator_, X_test, Y_test)

grid_rf = GridSearchCV(pipe_rf, rf_params, cv=5, scoring='accuracy')
grid_rf.fit(X_train, Y_train)
print("Best RF Parameters:", grid_rf.best_params_)
evaluate_model(grid_rf.best_estimator_, X_test, Y_test)

grid_svm = GridSearchCV(pipe_svm, svm_params, cv=5, scoring='accuracy')
grid_svm.fit(X_train, Y_train)
print("Best SVM Parameters:", grid_svm.best_params_)
evaluate_model(grid_svm.best_estimator_, X_test, Y_test)

grid_cat = GridSearchCV(pipe_cat, cat_params, cv=5, scoring='accuracy')
grid_cat.fit(X_train, Y_train)
print("Best CatBoost Parameters:", grid_cat.best_params_)
evaluate_model(grid_cat.best_estimator_, X_test, Y_test)
```

```python
# Make sure you have a copy of original features
X_orig = df.drop('Outcome', axis=1)

# Random Forest Feature Importance
importances = grid_rf.best_estimator_.named_steps['classifier'].feature_importances_
feature_names = X_orig.columns  # Use original DataFrame columns

# Sort for plotting
sorted_idx = importances.argsort()
plt.barh(feature_names[sorted_idx], importances[sorted_idx])
plt.title("Feature Importance (Random Forest)")
plt.xlabel("Importance Score")
plt.show()

# Logistic Regression Coefficients
coef = pd.Series(grid_lr.best_estimator_.named_steps['classifier'].coef_[0], index=X_orig.columns)
coef = coef.sort_values()
coef.plot(kind='barh', color='skyblue')
plt.title("Feature Importance (Logistic Regression)")
plt.xlabel("Coefficient Value")
plt.show()


```

```python
# Step 11: Compare Models

models = {
    "Logistic Regression": grid_lr.best_estimator_,
    "Random Forest": grid_rf.best_estimator_,
    "SVM": grid_svm.best_estimator_,
    "CatBoost": grid_cat.best_estimator_
}

summary = []

for name, model in models.items():
    y_pred = model.predict(X_test)
    if hasattr(model, "predict_proba"):
        y_prob = model.predict_proba(X_test)[:,1]
    else:
        y_prob = model.decision_function(X_test)
        y_prob = (y_prob - y_prob.min()) / (y_prob.max() - y_prob.min())
    
    acc = accuracy_score(Y_test, y_pred)
    auc = roc_auc_score(Y_test, y_prob)
    report = classification_report(Y_test, y_pred, output_dict=True)
    recall = report['1']['recall']  # recall of diabetic patients (minority class)
    
    summary.append([name, acc, recall, auc])

# Convert to DataFrame
summary_df = pd.DataFrame(summary, columns=['Model', 'Accuracy', 'Recall (Class 1)', 'ROC-AUC'])
print(summary_df.sort_values(by='Recall (Class 1)', ascending=False))
```
